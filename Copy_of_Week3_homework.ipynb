````python
# # Find the mean and standard deviation of glycated haemoglobin and glucose in healthy patients and in patients with diabetes.
# automate the subsetting 
diabetes_yes_no = {0: 'healthy patients', 1: 'diabetic patients'}
for i in range(0,2):
  diabetic_subset = alldata[alldata['Diabetes diagnosed by doctor'] == i]
  for biomarker in ['Glucose [mmol/L]', 'Glycated haemoglobin (HbA1c) [mmol/mol]']:
    mean = np.mean(diabetic_subset[biomarker])
    std = np.std(diabetic_subset[biomarker])
    print('In {}, {} mean is {:.3f} ± {:.3f}'.format(diabetes_yes_no[i], biomarker, mean, std ))
````

````python
# # Divide the population by age, and find mean and standard deviation of glycated haemoglobin and glucose in healthy patients and in patients with diabetes in the different age groups (example: younger than 50, 50-54, 55-59, 60-64, 65-69, older than 70)
````

````python
#Elisa ansawer

conditions = [alldata['AgeRecruit'] < 50, 
              (alldata['AgeRecruit'] >= 50) & (alldata['AgeRecruit'] < 55),
              (alldata['AgeRecruit'] >= 55) & (alldata['AgeRecruit'] < 60),
              (alldata['AgeRecruit'] >= 60) & (alldata['AgeRecruit'] < 65),
              alldata['AgeRecruit'] >= 65
              ]

choices = [0, 1, 2, 3, 4]

alldata['AgeBin'] = np.select(conditions, choices, default = np.nan)

age_bin_dic = {0 : 'younger than 50',
               1: "between 50 and 54",
               2: "between 55 and 60",
               3: "between 60 and 64",
               4: 'older than 65'}

diabetes_yes_no = {0: 'patients without diabetes', 1: 'patients with diabetes'}

for age in age_bin_dic:
  age_subset = alldata[alldata["AgeBin"] == age]

  for i in range(0,2):
    subset = age_subset[age_subset['Diabetes diagnosed by doctor'] == i]
    for b in ['Glucose [mmol/L]', 'Glycated haemoglobin (HbA1c) [mmol/mol]']:
      mean = np.mean(subset[biomarker])
      std = np.std(subset[biomarker])
      print('In {} {}, {} mean is {:.3f} and standard deviation is {:.3f}'.format(diabetes_yes_no[i], age_bin_dic[age],  b, mean, std ))
````
